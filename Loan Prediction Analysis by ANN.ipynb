Creating the "replacer" function for Missing Data Treatment in which null values of numerical data will be replaced by mean of numerical data and categorical data null values will be replaced by mode of categorical data present in the dataset.

```python
def replacer(df):
    import pandas as pd
    Q = pd.DataFrame(df.isna().sum(),columns=["ct"])
    for i in Q[Q.ct > 0].index:
        if(df[i].dtypes == "object"):
            x = df[i].mode()[0]
            df[i] = df[i].fillna(x)
        else:
            x = df[i].mean()
            df[i] = df[i].fillna(x)
```

```python
replacer(B)
```

```python
Y = B[["Loan_Status"]]
X = B.drop(labels=["Loan_Status"],axis=1)
```

```python
X
```

The important columns need to have a same magnitude before creating the model so we will create another new function know as Preprocessing for the DataSet.

```python
def preprocessing(X):
    cat = []
    con = []
    import pandas as pd
    for i in X.columns:
        if(X[i].dtypes=="object"):
            cat.append(i)
        else:
            con.append(i)      

    X1 = pd.get_dummies(X[cat])
    from sklearn.preprocessing import StandardScaler
    ss = StandardScaler()
    X2 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
    Xnew = X2.join(X1)
    return Xnew
```

```python
Xnew = preprocessing(X)
```

```python
Xnew
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y.Loan_Status = le.fit_transform(Y)
```

```python
Y
```

# Split the DataSet in Training and Testing Set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

# Creat an Artificial Neural Network(ANN)

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
nn = Sequential()
nn.add(Dense(200,input_dim=(Xnew.shape[1])))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(1,activation="sigmoid"))
```

```python
nn.compile(loss="binary_crossentropy",metrics=["accuracy"])
nn.fit(xtrain,ytrain,validation_split=0.2,batch_size=4,epochs=100)
```
